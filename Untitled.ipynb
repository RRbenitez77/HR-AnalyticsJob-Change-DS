```python
X =['city', 'city_development_index', 'gender',
       'relevent_experience', 'enrolled_university', 'education_level',
       'major_discipline', 'experience', 'company_size', 'company_type',
       'last_new_job', 'training_hours']

cat_features=['city','gender','relevent_experience','enrolled_university','education_level','major_discipline',
                   'experience','company_size','company_type','last_new_job']
y=['target']
```

```python
df[cat_features] = df[cat_features].fillna('no_info')
```

```python
from catboost import CatBoostClassifier
from catboost import cv, Pool
```

```python
train_data = Pool(data=df[X],
                  label=df[y],
                  cat_features=cat_features
                 )

```

```python
params = {'cat_features': cat_features,
              'eval_metric': 'AUC',
          'loss_function': 'Logloss',
              'learning_rate': 0.1,
              'random_seed': 2007,
              'verbose':100}
```

```python
cv_data = cv(
    params = params,
    pool = train_data,
    fold_count=3,
    shuffle=True,
    partition_random_seed=0,
    stratified=True,
    verbose=False,
    early_stopping_rounds=200
)
```

```python
cv_data[cv_data['test-AUC-mean'] == cv_data['test-AUC-mean'].max()]
```

```python
params = {'iterations': 203,
    'cat_features': cat_features,
              'eval_metric': 'AUC',
          'loss_function': 'Logloss',
              'learning_rate': 0.1,
              'random_seed': 2007,
              'verbose':100}
```

```python
model = CatBoostClassifier(**params)
```

```python
model.fit(train_data)
```

```python
import shap
shap.initjs()
```

```python
explainer = shap.TreeExplainer(model)
shap_values = explainer.shap_values(df[X])
```

```python
shap.force_plot(explainer.expected_value, shap_values[0,:], df[X].iloc[0,:],link="logit")
```

```python
df['y_score'] = model.predict_proba(train_data)[:,1]
```
